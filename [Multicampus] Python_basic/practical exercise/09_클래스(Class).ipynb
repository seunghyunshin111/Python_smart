#### 클래스를 만들어보자

```python
class Bicycle():
    #초기화 함수 필수
    def __init__(self, wheel_size=100, body_color='white'):
        self.wheel = wheel_size
        self.color = body_color
    
    #필요한 함수들 선언
    def move(self, speed):
        print('시속 {}km로 이동'.format(speed))    

    def turn(self, direction):
        print('{}방향으로 회전'.format(direction))

    def stop(self):
        print('정지')
```

#### 클래스를 이용해 객체(인스턴스)를 만들어보자

**클래스를 호출 시(=인스턴스를 생성 시) 초기화함수가 호출된다.**

```python
samcheonri = Bicycle(200, 'yellow')
print('wheel_size:{}, body_color:{}'.format(samcheonri.wheel,
                                           samcheonri.color))
```

```python
samcheonri.move(100)
samcheonri.turn('우측')
samcheonri.stop()
```

### 클래스 내의 변수의 종류
1) 클래스 변수 :  초기화함수 밖에서 선언/ 클래스명.변수명  
2) 인스턴스 변수 :  초기화함수 내에서 선언/ 인스턴스명.변수명

```python
class Car():
    
    instance_count = 0 #클래스 변수
    
    def __init__(self, wheel, color):
        self.wheel = wheel
        self.color = color
        
        Car.instance_count += 1
        print('{}번째 인스턴스 생성'.format(Car.instance_count))
        
    def move(self, speed):
        pass
```

### 클래스 내의 함수(메서드)의 종류 : 인스턴스 함수, 클래스 함수, 정적(스태틱) 함수
1) 인스턴스 함수(메서드): 객체(인스턴스)의 기능  
2) 클래스 함수(메서드): 클래스 내 관리  
3) 정적(스태틱) 함수(메서드): 사실 상 클래스와 무관/관리의 차원  

```python
class Car2():
    
    instance_count = 0
    
    def __init__(self, wheel, color, model_code):
        self.wheel = wheel
        self.color = color
        self.model_code = model_code
        Car2.instance_count += 1

    #인스턴스 메서드(함수)
    def move(self, speed):
        print('시속 {}km로 이동'.format(speed))
    
    #인스턴스 메서드(함수)
    def stop(self):
        print('정지!')
    
    #클래스 메서드(함수)
    @classmethod
    def count_instance(cls):
        print('현재 인스턴스 수 :{}'.format(Car2.instance_count))
    
    @staticmethod
    def check_type(model_code):
        if model_code >=20:
            print('전기차입니다.')
        elif model_code >= 10:
            print('수소차입니다.')
        else:
            print('디젤차입니다.')
    
```

### 상속

#### 기존의 클래스와 일부 겹치는 요소를 갖는 새로운 클래스를 만들 때, 겹치는 부분을 그대로 가져다 쓸 수 있으면 편하지 않을까?

코드를 재사용하기 위해서 만들어져있는 클래스를 상속받아 새로운 클래스를 만들 수 있다.
- class 클래스명(부모클래스명):
- 부모클래스명 default : Object 

```python
class Bicycle():
    #초기화 함수 필수
    def __init__(self, wheel_size=100, body_color='white'):
        self.wheel = wheel_size
        self.color = body_color
    
    #필요한 함수들 선언
    def move(self, speed):
        print('시속 {}km로 이동'.format(speed))    

    def turn(self, direction):
        print('{}방향으로 회전'.format(direction))

    def stop(self):
        print('정지')
```

```python
class FoldingBicycle(Bicycle):
    
    def __init__(self, wheel_size, body_color, folding_state):
        super().__init__(wheel_size, body_color)
        self.state = folding_state
    
    def move(self, speed, direction):
        super().move(speed)
        super().turn(direction)
```

```python
minivelo = FoldingBicycle(100, 'white', 'fold')
minivelo.wheel
```

```python
minivelo.stop()
```

```python
minivelo.move(100, 'right')
```

```python
minivelo.color
```

```python
minivelo.move(10, '좌측')
```

### 제너레이터

- 제너레이터는 반복자다.
- 메모리를 사용하지 않기 때문에, 더 좋은 퍼포먼스를 낸다.

```python
def operator_m3(nums):
    print(nums)
    result = []
    for num in nums:
        result.append(num*3)
        print(num)
        print(result)
    return result
    

numbers = [1, 2, 3]

result = operator_m3(numbers)
result
```

```python
#제너레이터
def operator_m3(nums):
    for num in nums:
        yield num*3
    

numbers = [1, 2, 3]

result = operator_m3(numbers)
result   
```

```python
print(next(result)) #1
```

```python
print(next(result)) #2
```

```python
print(next(result)) #3
```

```python
def operator_m3(nums):
    for num in nums:
        yield num*3
    

numbers = [1, 2, 3]

result = operator_m3(numbers)

for r in result:
    print(r)
```
