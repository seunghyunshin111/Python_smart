```
url = 'https://search.naver.com/search.naver?where=news&query=맨시티'

resp = requests.get(url)

soup = BeautifulSoup(resp.content, 'lxml')
a_tag = soup.select('ul.type01 li dl dt a')

title = []
for a in a_tag:
  title.append(a.text)

return title
```

정답코드

```
#keyword = 맨시티 
def news_crawling(message):
  
  url = 'https://search.naver.com/search.naver?where=news&query=' + message #문자열 + 문자열

  resp = requests.get(url)

  soup = BeautifulSoup(resp.content, 'lxml')
  a_tag = soup.select('ul.type01 li dl dt a')

  title = []
  for a in a_tag:
    title.append(a.text)

  print(title)
  return title, url
```

#### 3번

아래 두 함수를 이용해서, 원하는 만큼 숫자를 입력받은 num_list의 원소를 절대값 크기 순으로 정렬한 sorted_list 를 만들어보세요.

예시. 
num_list = [1, -2, 3, 4, 0]  

sorted_list = sort_data(num_list)  
sorted_list  
출력 결과 :  
  [4, 3, -2, 1, 0]

```
# 입력받은 데이터를 절대값으로 변환해주는 함수
def to_abs(num):
  if num >= 0:
    return num
  else:
    return -num
```

```
#  입력받은 데이터를 정렬해주는 함수
def sort_data(x_list):
  sort_list = []
  for idx, x in enumerate(x_list):
    check = 0
    if idx == 0:
      sort_list.append(x)
      continue
    for sort in sort_list:
      if to_abs(x) >= to_abs(sort):
        sort_list.insert(sort_list.index(sort), x)
        check = 1
        break
    if check == 0:
      sort_list.append(x)
    
  return sort_list
```

```
length = int(input('몇 개의 숫자를 입력하시겠습니까?'))

num_list = [int(input('입력 {} -> '.format(i))) for i in range(length)]
num_list
```

```
sorted_list = sort_data(num_list)
sorted_list
```

#### 4번

가변 매개변수를 이용해 입력받은 수의 평균을 구하는 average() 함수와 합을 구하는 sum() 함수를 구현하세요.

예시, 
print(average(1, 2, 3))  
print(sum(1, 2, 3))  

출력 결과  
2  
6

```
def my_average(*data):
  result, length = 0, len(data)
  # print(data)
  for d in data:
    result += d/length
  return result

def my_sum(*data):
  result, length = 0, len(data)
  for d in data:
    result += d
  return result

print(my_average(*[1, 2, 3, 4, 5]))
print(my_sum(1, 2, 3, 4, 5))
```

리스트와 인덱스를 찾고 싶은 값을 모두 입력하면 입력한 값의 인덱스(위치)를 리턴해주는 함수 만들기  
  
실행:

```
def find_index(nums, target):
  indexs = []
  for idx, num in enumerate(nums):
    if num == target:
      indexs.append(idx)
  return indexs


num_list=[1,2,3,1,4,2,1]

find_index(num_list, 5)
```
