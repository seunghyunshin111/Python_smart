#### 1번
사용자로점수를3개입력받아 모든점수가65점보다클경우True 점수가하나라도65점이안넘을경우False 를출력하세요

```
score = []
for _ in range(3):
  score.append(input('점수 입력:'))

check = 0
for scr in score:
  if scr < 65:
    check = 1
    break

if check == 0:
  print('True')
else:
  print('False')
```

#### 3번
사용자로점수를3개입력받으세요.  
모든점수가 65점보다 클 경우 -> 합격  
아닐경우 -> 불합격  
을출력하세요  
단,0~100이아닌숫자가입력된경우  
잘못된“잘못된점수가 입력되었습니다"를출력하세요


```
a = input("첫번째 점수를 입력해주세요 : ")
b = input("두번째 점수를 입력해주세요 : ")
c = input("세번째 점수를 입력해주세요 : ")

if (not(0<=int(a)<=100) or not(0<=int(b)<=100) or not(0<=int(c)<=100)):
    print("잘못된 점수가 입력되었습니다")
elif (int(a) > 65 and int(b) > 65 and int(c) > 65):
    print("합격")
else:
    print("불합격")
```

#### 5번
숫자를 입력받고 입력받은 정수의 구구단을 출력하세요.  
정수가 아닌 다른 타입이 입력되면,  
'정수를 입력해주세요!' 출력   
2-9가 아닌 수를 입력하면,  
" 2-9 사이의 정수를 입력해주세요!" 출력
  
실행 :  
몇단을 출력하시겠습니까? 3   
3 * 1 = 3   
3 * 2 = 6   
...   
3 * 8 = 24  
3 * 9 = 27

```
dan = input('몇 단을 출력하시겠습니까?')

check_int = 0
for dn in dan:
  if int(dn) in range(10):
    continue
  else:
    print('정수를 입력해주세요!') 
    check_int = 1 
    break 

if check_int == 0:
  dan = int(dan)
  if 2 <= dan <= 9: 
    for num in range(1, 10):
        print("{} * {} = {}".format(dan, num, dan*num))
  else:
    print('2-9 사이의 정수를 입력해주세요!')
```

#### 6번
사용자로부터 숫자를 계속 입력받다가  0을 입력하면 합계를 출력해주세요  
  
  실행 :  
값을 입력해주세요 : 30   
값을 입력해주세요 : 20   
값을 입력해주세요 : 50   
값을 입력해주세요 : 40   
값을 입력해주세요 : 30   
값을 입력해주세요 : 0   
합계는 ?  170

```
i = ''
sum_ = 0

while i != 0:
    i = int(input("값을 입력해주세요 : "))
    sum_ += i

print("합계는 ? ", sum_)
```
