### 웹툰데이터 관리하기

```
# 아래 코드는 네이버 웹툰페이지에서 월요일 웹툰의 제목/작가/평점을 가져오는 크롤링 코드입니다.
# 크롤링 과정에서 실수로 인해 작가 정보에 일부 '전체보기'라는 잘못된 정보가 들어가 있습니다. 
# 그럼 코드와 함께 문제를 확인하세요.
import requests
from bs4 import BeautifulSoup

url = 'https://comic.naver.com/webtoon/weekdayList.nhn?week=mon'

resp = requests.get(url)

soup = BeautifulSoup(resp.content, 'lxml')
title_tag = soup.select('ul.img_list li dl dt a')
author_tag = soup.select('ul.img_list li dl dd a')
rating_tag = soup.select('ul.img_list li dl dd strong')

# 크롤링 결과 확인
# print(len(rating_tag)) 
# print(title_tag)

# 크롤링한 정보를 저장
data_list = []
for title, author, rating in zip(title_tag, author_tag, rating_tag):
    data = {
        'title' : title.text,
        'author' : author.text,
        'rating' : rating.text,
    }
    data_list.append(data)
len(data_list), data_list
```

#### 1번
author가 전체보기로 되어있는 데이터는 data_list에서 모두 삭제해주세요.

```
#1. while 문 이용
webtoon_list = []

while data_list:

  data = data_list.pop(0)
  if data['author'] != '전체보기':
      webtoon_list.append(data)
len(webtoon_list), webtoon_list
```

```
#2. for-리스트 문 이용
webtoon_list = []

for data in data_list: #webtoon = webtoon_list[0]->webtoon_list[1]->webtoon_list[2]...->webtoon_list[-1]
  if data['author'] != '전체보기':
    webtoon_list.append(data)

del data_list
```

```
len(webtoon_list), webtoon_list
```

```
#3. 깊은복사 + for-리스트 문 이용
from copy import deepcopy # 깊은 복사 

webtoon_list = deepcopy(data_list)

for data in data_list:
    if data['author'] == '전체보기':
        webtoon_list.remove(data) 

del data_list

len(webtoon_list), webtoon_list
```

#### 2번
(author가 제대로 들어가있는 웹툰들 중) rating이 9.9 이상인 웹툰의 제목 따로 recommended 변수에 담아주세요.

```
recommended = []

for webtoon in webtoon_list:
  if float(webtoon['rating']) >= 9.9:
    recommended.append(webtoon['title'])
```

```
recommended
```

#### 3번
(author가 제대로 들어가있는 웹툰들 중) rating이 제일 높은 데이터 3개만 top_recommended에 담아주세요.

```
top_recommended = []

for webtoon in webtoon_list:
    if not top_recommended:
        top_recommended.append(webtoon)
        continue
        
    for top in top_recommended:
        if webtoon['rating'] >= top['rating']:
            top_recommended.insert(top_recommended.index(top), webtoon)
            if len(top_recommended) > 3: 
                top_recommended = top_recommended[:3]
            break 
      

top_recommended
```
