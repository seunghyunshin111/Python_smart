```python
# x축, y축 모두 있는 데이터 => 2차원 선 그래프

import numpy as np

x = np.arange(-4.5, 5, 0.5)  # 배열 x 생성. 범위: 0.5싹 증가
y = 2 * x ** 2  # 수식을 이용해 배열 x에 대응하는 배열 y 생성

[x, y]
```

```python
# 여러 그래프 그리기

import numpy as np

x = np.arange(-4.5, 5, 0.5)
y1 = 2 * x**2
y2 = 5 * x + 30
y3 = 4 * x**2 + 10
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

```python
# 새로운 그래프 창에 그래프를 그리려면

plt.plot(x, y1)

plt.figure()   # 새로운 그래프 창을 생성함
plt.plot(x,y2) # 새롭게 생성된 그래프 창에 그래프를 그림

plt.show()
```

### 그래프 창의 번호를 명시적으로 지정한 후, 해당 창에 그래프를 그릴 수도 있다.
### 인자 n(정수)를 지정하면 지정된 번호로 그래프 창이 지정된다.
### 지정된 번호의 그래프 창이 없다면, 새로 그래프 창을 생성한 후에 그래프가 그려진다!

새로운 창에는 제목에 'Figure 번호'가 표시되지만,
inline에서 실행된 경우는 번호가 명시적으로 나타나지는 않는다.

## 그래프 삭제 clf()
## 현재 그래프 창 닫기 close()

```python
import numpy as np

# 데이터 생성

x = np.arange(-5, 5, 0.1)
y1 = x**2 - 2
y2 = 20 * np.cos(x)**2    # NumPy에서 cos()는 np.cos()로 입력

plt.figure(1)     # 1번 그래프 창을 생성함
plt.plot(x, y1)   # 지정된 그래프 창에 그래프를 그림

plt.figure(2)     # 2번 그래프 창을 생성함
plt.plot(x, y2)   # 지정된 그래프 창에 그래프를 그림

plt.figure(1)     # 이미 생성된 1번 그래프 창을 지정함
plt.plot(x, y2)   # 지정된 그래프 창에 그래프를 그림

plt.figure(2)     # 이미 생성된 2번 그래프 창을 지정함
plt.clf()         # 2번 그래프 창에 그려진 모든 그래프를 지움
plt.plot(x, y1)   # 지정된 그래프 창에 그래프를 그림

plt.show()
```

```python
import numpy as np

# 데이터 생성

x = np.arange(-5, 5, 0.1)
y1 = x**2 - 2
y2 = 20 * np.cos(x)**2    # NumPy에서 cos()는 np.cos()로 입력

plt.figure(1)     # 1번 그래프 창을 생성함
plt.plot(x, y1)   # 지정된 그래프 창에 그래프를 그림
```

```python
plt.figure(2)     # 2번 그래프 창을 생성함
plt.plot(x, y2)   # 지정된 그래프 창에 그래프를 그림
```

```python
plt.figure(1)     # 이미 생성된 1번 그래프 창을 지정함
plt.plot(x, y2)   # 지정된 그래프 창에 그래프를 그림
```

```python
plt.figure(2)     # 이미 생성된 2번 그래프 창을 지정함
plt.clf()         # 2번 그래프 창에 그려진 모든 그래프를 지움
plt.plot(x, y1)   # 지정된 그래프 창에 그래프를 그림
```
