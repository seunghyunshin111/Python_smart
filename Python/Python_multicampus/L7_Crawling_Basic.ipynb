연관 검색어를 가져와보세요.

```
import requests
from bs4 import BeautifulSoup

url = 'https://search.naver.com/search.naver' # 프로토콜 + 주소값 + 리소스경로
# query = '?where=nexearch&query=%EC%82%BC%EB%8B%A4%EC%88%98'
query = {
    'where' : 'nexearch',
    'query' : '삼다수'
} # 쿼리스트링

resp = requests.get(url, query)
resp.status_code
```

파싱 =  html에서 원하는 tag를 가져오는 행위  
파서 = lxml / html5lib / html.parser  
3개 중에 lxml이 가장 빠르고, 효율적이다.

```
#html에서 무엇을? = tag = 데이터 == 연관검색어
soup = BeautifulSoup(resp.content, 'lxml') # 파서 = lxml
# html = resp.content = soup
a_tag = soup.select('ul._related_keyword_ul li a') #이러한 a태그를 전부 다 가져와
len(a_tag) #a_tag는 리스트다.
```

```
#많이 하는 실수 
# a_tag.text # a_tag는 리스트다. 리스트에다가 .text를 사용할 수는 없다.
# a_tag[0].text # .text는 태그에만 사용할 수 있는 beautifulsoup 기능이다.
# a_tag[1].text
# a_tag[2].text
# ...
# a_tag[9].text

related = []
for a in a_tag: # a <- 첫번째 a태그부터 마지막 a태그까지 
  related.append(a.text)

related
```

뉴스토픽을 가져와 보세요.
- 뉴스 와 연예/스포츠를 구분지어서 가져와보세요.

```
#1단계(어디서?)는 위의 연관검색어 동일하다.
import requests
from bs4 import BeautifulSoup

url = 'https://search.naver.com/search.naver' # 프로토콜 + 주소값 + 리소스경로
# query = '?where=nexearch&query=%EC%82%BC%EB%8B%A4%EC%88%98'
query = {
    'where' : 'nexearch',
    'query' : '삼다수'
} # 쿼리스트링

resp = requests.get(url, query)
resp.status_code
```

속성 값에 ' '(공백, 스페이스)는 .(점)으로 대체해준다.  
ol.lst_realtime_srch _tab_area  
->  
ol.lst_realtime_srch._tab_area  


```
#2단계(무엇을?) == tag = 데이터 == 뉴스토픽
# ol.lst_realtime_srch _tab_area li a span span
soup = BeautifulSoup(resp.content, 'lxml')
ol_tag = soup.select('ol.lst_realtime_srch._tab_area')
len(ol_tag)
```

첫번째 ol 태그에는 뉴스 10개  
두번째 ol 태그에는 연애.스포츠 10개

```
# span_tag_0 = ol_tag[0].select('li a span span')
# span_tag_1 = ol_tag[1].select('li a span span')

topics = []
for ol in ol_tag: # 2바퀴
  topic = []
  span_tag = ol.select('li a span span')
  for span in span_tag: # 10바퀴
    topic.append(span.text) #topic 에는 10개의 뉴스토픽이 들어간다.
  topics.append(topic)
topics
```

뉴스 기사 제목과 링크를 100개 가져와서 저장해보세요. 

```
# page = 1, 2, 3, ~~~ 10
no_page = 10

news_title = []
news_link = []
for page in range(no_page): 
  url = 'https://search.naver.com/search.naver'
  # 'where=news&query=깡'
  query = {
      'where' : 'news',
      'query' : '깡',
      'start' : page*10 + 1   
  } # start : 1, 11, 21, 31, ~~~ 91

  resp = requests.get(url, query)

  soup = BeautifulSoup(resp.content, 'lxml')
  #ul.type01 li dl dt a
  a_tag = soup.select('ul.type01 li dl dt a')
  

  for a in a_tag: # 10바퀴 * 10바퀴 = 100바퀴
    news_title.append(a['title'])
    news_link.append(a['href'])
len(news_title), len(news_link)
```

이 src는 바로, 이미지가 저장되어있는 url이다!!!  
  
  해당 src (url)에 requests.get을 요청하면 된다.


이미지 크롤링 절차  
1) 어디서? = url = requests.get  
2) 무엇을? = tag = soup.select = img의 src   
3) 어디서? = src url  
4) 무엇을? 이미지

```
url = 'https://www.python.org'
resp = requests.get(url)

soup = BeautifulSoup(resp.content, 'lxml')
img_tag = soup.select('img.python-logo')
# img_tag.src # 리스트는 .src를 할 수 없다.
src = img_tag[0]['src'] # == img_tag[0].src
src
```

```
src_url = url + src
src_url
```

```
resp = requests.get(src_url)
resp.content # 인코딩된 img 파일
```
