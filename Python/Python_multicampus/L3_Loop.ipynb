#### for i in 리스트 용법

- i 는 리스트의 원소다. (i am list)
- 리스트의 길이만큼 돌고, 마지막 원소까지 돈다.
- 단점: 몇번째 loop인지를 어떻게 알지?

```python
cafe = ['스타벅스', '탐앤탐스', '투썸플레이스', '이디야', '할리스']
for i in cafe:
    print(i)
```

```python
cafe = ['스타벅스', '탐앤탐스', '투썸플레이스', '이디야', '할리스']
for i in cafe:
  print(i, end = " ")
```

##### for idx, i in enumerate(리스트)

```python
for idx, cf in enumerate(cafe):
    print(idx, cf)
```

```python
for idx, cf in enumerate(cafe):
  print(idx, cf, end = " ")
```

##### for i, j in zip(리스트1, 리스트2)

```python
title = ['신의탑', '헬퍼', '외모지상주의', '호랑이형님']
rating = ['9.94', '9.90', '9.46']#갯수 적은 것에 맞춤


for tt, rt in zip(title, rating):#f=food[0], c=coffee[0]
  print(tt, rt)
```

```python
title = ['신의탑', '헬퍼', '소녀의 세계', '호랑이형님']
rating = ['9.94', '9.90', '9.46']#갯수 적은 것에 맞춤

for ti, ra in zip(title, rating):
  print(ra, ti, sep = " ")
  print(ti, ra)
```

## While

#### while(조건) : 조건이 참인 동안 loop

##### 조건에 대한 고찰

- 일반적인 조건문: 비교연산식 ex) num > 5, movie == 'ones'
- 비교연산식 + 논리연산 ex) num > 5 and movie != 'ones'

- 다른 조건문  
1) bool data ex) True, False  
2) number data ex) 1, 0 심지어 10, 100, -1  
3) string data ex) ' ', '', 'a', 그외의 모든 string 

```python
if True:
    print('hi')
```

```python
if 1:
    print('hi 1')

if 0:                  # 0: False
    print('hi 0')

if -1:
    print('hi -1')

if 1000:
    print('hi 1000')
```

- 숫자의 경우 : 0 만 False (0이외의 수는 모두 True)

```python
if ' ':
    print('hi space')
    
if '':               # "": (null) == False
    print('hi null')
    
if '0':
    print('hi "0"')

if 'spider-man':
    print('hi spider-man')
```

- 문자의 경우 : ''(null)만 False

##### while true ~ break 용법

```python
#무한반복
count = 0
while True:
    count += 1 #-> count = count + 1
    if count >= 100:
        print('loop finished')
        break #바로 상위의 loop 1개 종료
    print(count)


# 100 출력 안 됨
```

##### continue 용법

```python
# 1~10까지의 합을 구할 경우, 5의 배수는 더하고 싶지 않을 때
# 무한루프를 이용해보자.
```

```python
#직접풀어보세요

sum = 0
num = 0

while 1:
  if num == 10:
    break

  num += 1


  if num%5 == 0:
    continue

  sum += num

sum
```

## 중첩 Loop (다중 Loop)

```python
count = 0 
print('start:', count)
for i in range(10):
    count += 1
    for j in range(5):
        count -= 1
        print(count)
    print(i, '번째 loop result:', count) 

print('result:', count) 
```

```python
count = 0 

for i in range(10):
  for j in range(5):
    if i != j:
      count += 1
    else:
      count = 0

count
```

```python
x = [3, 6, 9, 20, -7, 5]

for i in x:
  if i < 0:
    i = i * -1
     
  if i % 2 == 0:
      i = i * 10
  elif i % 2 != 0:
      i = i * 10

  print(i)
```
