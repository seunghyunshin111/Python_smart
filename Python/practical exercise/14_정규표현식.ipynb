- 스페이스 뒤, 점 앞에 있는 문자열
-  (문자열). 
- \s(.+?)\.

```
re.findall('\s(.+?)\.', Name)
```

### Greedy Q VS Reluctant Q

.+ 와 .+?는 모두 '문자열'이란 뜻으로 해석하면 된다. 
- 다른 부연설명 없이 문자열이라는 표현

```
Text = '''멀티캠퍼스(Multicampus)는 국내외 17,000여 개의 기업에 서비스(Service)를 제공하고 있다.'''
```

```
# 소괄호 안에 감싸져 있는 문자열
# = \((문자열)\)
# = \((.+)\) 냐 아니면 \((.+?)\) 냐 
# Greedy 를 쓸 것이냐, Reluctant 를 쓸 것이냐.
# 욕심쟁이냐 귀차니즘이냐
```

```
re.findall('\((.+)\)', Text)
```

```
re.findall('\((.+?)\)', Text)
```
