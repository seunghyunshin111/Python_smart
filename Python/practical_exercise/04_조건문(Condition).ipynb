# 조건문(Condition)

특정 조건을 만족할 때만 수행되었으면 하는 코드가 있다면?

## If A: 만약에 A라면

예시)
보고 싶은 영화의 잔여석이 0이 아니라면 티켓 요금을 안내해주는 문구를 띄워주는 코드를 만들어보자.

```python
movie_info = {'기생충' : {'잔여석' : 28,
                        '티켓요금' : 11000}}
```

```python
my_movie = '기생충'

movie_info[my_movie]['잔여석']
```

```python
seat = movie_info[my_movie]['잔여석']
fare = movie_info[my_movie]['티켓요금']

if seat != 0: #if (조건): = (조건)이 참이라면
    print('예매 도와드리겠습니다.')
    print('티켓요금은', fare, '입니다.')
```

예시)

잔여석이 0이라면 죄송하다는 문구를 띄워주는 코드를 만들어보자.

## else : 그렇지 않다면

```python
movie_info = {'기생충' : {'잔여석' : 28,
                        '티켓요금' : 11000},
              '밀정' :  {'잔여석' : 0,
                        '티켓요금' : 11000}}
```

```python
# my_movie = '기생충'
my_movie = '밀정'
```

```python
seat = movie_info[my_movie]['잔여석']
fare = movie_info[my_movie]['티켓요금']

if seat != 0: #if (조건): = (조건)이 참이라면
    print('예매 도와드리겠습니다.')
    print('티켓요금은', fare, '입니다.')
else:
    print('죄송합니다.')
    print('잔여석이 없습니다.')
```

## elif B: 그렇지않고 B라면

가장 보고 싶은 영화의 잔여석을 먼저 확인해서 잔여석이 있으면 티켓요금을 안내해주고,  
그렇지않고, 두번째 보고 싶은 영화의 잔여석은 있다면 두번째 영화의 티켓요금을 안내해주고,
그렇지도 않다면 죄송하다는 문구를 띄워주도록 코드를 작성해보자

```python
movie_info = {'기생충' : {'잔여석' : 28,
                        '티켓요금' : 11000},
              '밀정' :  {'잔여석' : 0,
                        '티켓요금' : 11000}}
```

```python
my_movie = ['밀정', '기생충']
```

```python
seat = [movie_info[my_movie[0]]['잔여석'], movie_info[my_movie[1]]['잔여석']] 
fare = [movie_info[my_movie[0]]['티켓요금'], movie_info[my_movie[1]]['티켓요금']]

if seat[0] != 0: #if (조건): = (조건)이 참이라면
  print(my_movie[0], '예매 도와드리겠습니다.')
  print('티켓요금은', fare[0], '입니다.')
elif seat[1] != 0:
  print('죄송합니다.')
  print(my_movie[0], '잔여석이 없습니다.')
  print('대신,' my_movie[1], '예매 도와드리겠습니다.')
  print('티켓요금은', fare[1], '입니다.')
else:
  print('죄송합니다.')
  print(my_movie[0], my_movie[1], '모두 잔여석이 없습니다.')  
```

먼저, 보고싶은 영화가 원하는 시간대에 있는지 확인하고, 있다면 잔여석을 확인해주는 코드를 작성해보세요.

## 응용 : 조건문 안에 조건문  
if A:  
-> if B:

```python
movie_info = {'기생충' :  {
                          '티켓요금' : 11000,
                          '7:00' : 28,
                          '10:00' : 0
                        },
              '밀정' :  {
                          '티켓요금' : 11000,
                          '7:30' : 23,
                          '10:30' : 102
                        }
              }
```

```python
'7:00' in movie_info['기생충']
```

```python
my_movie = '기생충'
my_time = '11:00'

if my_time in movie_info[my_movie]:
  print(my_movie, my_time, '영화 있습니다.')
  if movie_info[my_movie][my_time]:
    print(my_movie, '예매 도와드리겠습니다.')
    print('티켓요금은', movie_info[my_movie]['티켓요금'], '입니다.')
  else:
    print('죄송합니다.')
    print('잔여석이 없습니다.')
else:
  print('죄송합니다.')
  print(my_movie, '영화는', my_time, '시간대가 없습니다.')
```

취미가 여행이라면, 나라를 추천해주되, 다녀온적이 있는 나라는 제외하고 추천해주는 코드를 작성해보세요.

```python
per_info = {'id': 0,
            'hobby': 'travel', 
           'have_been': ['china', 'japan']}

country = ['japan', 'china', 'swiss']

import copy
country_list = copy.deepcopy(country)

if per_info['hobby'] == 'travel':
    if country[0] in per_info['have_been']:
        country_list.remove(country[0])
    if country[1] in per_info['have_been']:
        country_list.remove(country[1])
    if country[2] in per_info['have_been']:
        country_list.remove(country[2])
    print(country_list)
```
