#### for i in range(시작점, 끝점+1, 변화간격) 용법

#### for i in 리스트 용법

- i 는 리스트의 원소다. (i am list)
- 리스트의 길이만큼 돌고, 마지막 원소까지 돈다.
- 단점: 몇번째 loop인지를 어떻게 알지?

```python
cafe = ['스타벅스', '탐앤탐스', '투썸플레이스', '이디야', '할리스']
for i in cafe:
    print(i)
```

##### for idx, i in enumerate(리스트)

```python
for idx, cf in enumerate(cafe):
    print(idx, cf)
```

여러개의 리스트를 한번에 loop 시킬 수 있을까?

##### for i, j in zip(리스트1, 리스트2)

```python
title = ['신의탑', '헬퍼', '외모지상주의', '호랑이형님']
rating = ['9.94', '9.90', '9.46']#갯수 적은 것에 맞춤


for tt, rt in zip(title, rating):#f=food[0], c=coffee[0]
  print(tt, rt)
```

```python
#enumerate과 zip 함께 써보자.

title = ['신의탑', '헬퍼', '외모지상주의', '호랑이형님']
rating = ['9.94', '9.90', '9.46']#갯수 적은 것에 맞춤

for idx, (tt, rt) in enumerate(zip(title, rating)):#f=food[0], c=coffee[0]
  print(idx, tt, rt)
```

## While

#### while(조건) : 조건이 참인 동안 loop

##### 조건에 대한 고찰

- 일반적인 조건문: 비교연산식 ex) num > 5, movie == 'ones'
- 비교연산식 + 논리연산 ex) num > 5 and movie != 'ones'

- 다른 조건문  
1) bool data ex) True, False  
2) number data ex) 1, 0 심지어 10, 100, -1  
3) string data ex) ' ', '', 'a', 그외의 모든 string 

```python
if True:
    print('hi')
```

```python
if 1:
    print('hi 1')

if 0:
    print('hi 0')

if -1:
    print('hi -1')

if 1000:
    print('hi 1000')
```

- 숫자의 경우 : 0 만 False (0이외의 수는 모두 True)

```python
if ' ':
    print('hi space')
    
if '':
    print('hi null')
    
if '0':
    print('hi "0"')

if 'spider-man':
    print('hi spider-man')
```

- 문자의 경우 : ''(null)만 False

##### while true ~ break 용법

```python
#무한반복
count = 0
while True:
    count += 1 #-> count = count + 1
    if count >= 100:
        print('loop finished')
        break #바로 상위의 loop 1개 종료
    print(count)
```

##### continue 용법

```python
# 1~10까지의 합을 구할 경우, 5의 배수는 더하고 싶지 않을 때
# 무한루프를 이용해보자.

sum_, num = 0, 0

while 1:

#------------(1)---------------

  num += 1

  if num == 10:
    break

#------------(2)---------------

  if num%5 == 0:
    continue

  sum_ += num

#------------(3)---------------


#------------(4)---------------

sum_
```
