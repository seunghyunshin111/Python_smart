```python
#### 문제 1부터 100까지 숫자를 출력해보세요.  
# 조건1. 10단위로 한줄에 출력되도록 해보세요.  
# 결과 :  
# 1, 2, 3, ... , 10  
# 11, 12, ... , 20  
# ...  ... ...  ...  ...  
# 91, 92, ... ,100
```

```python
#풀이
for i in range(100):
  if (i+1)%10 == 0:
    print(i+1)
  else:
    print(i+1, end=' ')
```

```python
#1번
for i in range(100):
    if (i+1)%10 == 0: 
        print('{:3}'.format(i+1))
    else:
        print('{:3}'.format(i+1), end=' ')
```

```python
#2번
for i in range(100):
    if (i+1)%10 == 0: 
        print('{:03}'.format(i+1))
    else:
        print('{:03}'.format(i+1), end=' ')
```

```python
#3번
for i in range(100):
    if (i+1)%10 == 0: 
        print('{:<3}'.format(i+1))
    else:
        print('{:<3}'.format(i+1), end=' ')
```

#### 출력할 대상이 실수(float)일때는, 소숫점 자리가 문제다. 이 또한 형식을 지정해주려면? 

```python
print('{:.3f}'.format(13.333333333))
```

```python
for i in range(100):
    print('{:5.2f}'.format(i+1+0.356), end=' ')
    if (i+1)%10 == 0:
        print() #enter
```

#### 형식지정출력을 사용할 때, 자리를 지정해주려면?

```python
#자리 지정 가능
print('1st:{1}, 2nd:{2}, 3rd:{0}'.format('우디', '버즈', '포키'))
```

####(응용) 자리 지정과 자리수 지정 함께 사용하기  
결과 :  
1st:&nbsp;&nbsp;&nbsp;버즈 2nd:&nbsp;&nbsp;&nbsp;포키 3rd:&nbsp;&nbsp;&nbsp;우디  
1st:(스페이스3번)버즈 2nd:(스페이스3번)포키 3rd:(스페이스3번)우디

```python
#자리 지정과 자리수 지정 함께 쓰기
print('1st:{1:>5}, 2nd:{2:>5}, 3rd:{0:>5}'.format('우디', '버즈', '포키'))
```

```python
#자리 지정 가능
print('{a} x {b} = {mul}'.format(a=1, b=2, mul=1*2))
```
